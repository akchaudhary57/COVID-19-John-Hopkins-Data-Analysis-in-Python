```python
import pandas as pd
```

```python
covidWorldwide = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
covidWorldwide['continent'] = covidWorldwide['continent'].fillna(covidWorldwide['location'])
covidWorldwide['date'] = pd.to_datetime(covidWorldwide['date'])
```

```python
covidWorldwide.head()
```

```python
Recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
death = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
del Recovered['Lat']
del Recovered['Long']
del confirmed['Lat']
del confirmed['Long']
del death['Lat']
del death['Long']
```

```python
Recovered = Recovered.rename(columns = {'Province/State':'State','Country/Region':'Country'})
Recoveredtransform =Recovered.melt(id_vars = ['State','Country'])
Recoveredtransform = Recoveredtransform.rename(columns = {'variable':'Date','value':'Total_Recovered_Cases_by_state'})
Recoveredtransform['Date'] = pd.to_datetime(Recoveredtransform['Date'])
Recoveredtransform.columns = ['Recovered_State','Recovered_Country','Recovered_Date','Total_Recovered_Cases_by_state']
Recoveredtransform
```

```python
Confirmed = confirmed.rename(columns = {'Province/State':'State','Country/Region':'Country'})
Confirmedtransform =Confirmed.melt(id_vars = ['State','Country'])
Confirmedtransform = Confirmedtransform.rename(columns = {'variable':'Date','value':'Total_Confirmed_Cases_by_state'})
Confirmedtransform['Date'] = pd.to_datetime(Confirmedtransform['Date'])
Confirmedtransform.columns = ['Confirmed_State','Confirmed_Country','Confirmed_Date','Total_Confirmed_Cases_by_state']
Confirmedtransform
```

```python
death = death.rename(columns = {'Province/State':'State','Country/Region':'Country'})
deathtransform =death.melt(id_vars = ['State','Country'])
deathtransform = deathtransform.rename(columns = {'variable':'Date','value':'Total_death_Cases_by_state'})
deathtransform['Date'] = pd.to_datetime(deathtransform['Date'])
deathtransform.columns = ['death_State','death_Country','death_Date','Total_death_Cases_by_state']
deathtransform
```

```python
mergetransformcases = pd.concat([Recoveredtransform,Confirmedtransform,deathtransform],axis = 1)


mergetransformcases['Country'] = mergetransformcases['Recovered_Country'].fillna(mergetransformcases['Confirmed_Country'])
mergetransformcases['Country'] = mergetransformcases['Recovered_Country'].fillna(mergetransformcases['death_Country'])
mergetransformcases['Country'] = mergetransformcases['Confirmed_Country'].fillna(mergetransformcases['Recovered_Country'])
mergetransformcases['Country'] = mergetransformcases['Confirmed_Country'].fillna(mergetransformcases['death_Country'])
mergetransformcases['Country'] = mergetransformcases['death_Country'].fillna(mergetransformcases['Recovered_Country'])
mergetransformcases['Country'] = mergetransformcases['death_Country'].fillna(mergetransformcases['Confirmed_Country'])
mergetransformcases['State'] = mergetransformcases['Recovered_State'].fillna(mergetransformcases['Confirmed_State'])
mergetransformcases['State'] = mergetransformcases['Recovered_State'].fillna(mergetransformcases['death_State'])
mergetransformcases['State'] = mergetransformcases['Confirmed_State'].fillna(mergetransformcases['Recovered_State'])
mergetransformcases['State'] = mergetransformcases['Confirmed_State'].fillna(mergetransformcases['death_State'])
mergetransformcases['State'] = mergetransformcases['death_State'].fillna(mergetransformcases['Recovered_State'])
mergetransformcases['State'] = mergetransformcases['death_State'].fillna(mergetransformcases['Confirmed_State'])
mergetransformcases['Date'] = mergetransformcases['Recovered_Date'].fillna(mergetransformcases['Confirmed_Date'])
mergetransformcases['Date'] = mergetransformcases['Recovered_Date'].fillna(mergetransformcases['death_Date'])
mergetransformcases['Date'] = mergetransformcases['Confirmed_Date'].fillna(mergetransformcases['Recovered_Date'])
mergetransformcases['Date'] = mergetransformcases['Confirmed_Date'].fillna(mergetransformcases['death_Date'])
mergetransformcases['Date'] = mergetransformcases['death_Date'].fillna(mergetransformcases['Recovered_Date'])
mergetransformcases['Date'] = mergetransformcases['death_Date'].fillna(mergetransformcases['Confirmed_Date'])
mergetransformcases
```

```python
MergeFinal = mergetransformcases[['Country','State','Date','Total_Recovered_Cases_by_state','Total_Confirmed_Cases_by_state','Total_death_Cases_by_state']]
MergeFinal.head()
```

```python
mergecases = pd.merge(MergeFinal,covidWorldwide,how = 'outer',left_on = ['Country','Date'],right_on = ['location','date'])
mergecases.columns
```
