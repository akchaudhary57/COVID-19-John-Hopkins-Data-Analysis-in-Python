```python
import pandas as pd
```

```python
covidWorldwide = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
covidWorldwide['continent'] = covidWorldwide['continent'].fillna(covidWorldwide['location'])
```

```python
covidWorldwideAus = covidWorldwide[(covidWorldwide['date'] == '2020-07-24') & (covidWorldwide['location'] == 'Australia')]
```

```python
covidWorldwideAus['date'] = pd.to_datetime(covidWorldwideAus['date'])
```

```python
covidWorldwideAus.dtypes
```

```python
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
death = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
Recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
del confirmed['Lat']
del confirmed['Long']
```

```python
confirmed = confirmed.rename(columns = {'Province/State':'State','Country/Region':'Country'})
confirmedtransform =confirmed.melt(id_vars = ['State','Country'])
confirmedtransform = confirmedtransform.rename(columns = {'variable':'Date','value':'Cases'})
confirmedtransform['Date'] = pd.to_datetime(confirmedtransform['Date'])
confirmedtransformAus = confirmedtransform[(confirmedtransform['Date'] == '2020-07-24') & (confirmedtransform['Country'] == 'Australia')]
confirmedtransformAus.dtypes
```

```python
mergecases = pd.merge(confirmedtransformAus,covidWorldwideAus,how = 'outer',left_on = ['Country','Date'],right_on = ['location','date'])
mergecases
```
