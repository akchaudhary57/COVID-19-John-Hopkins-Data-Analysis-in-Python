### How to select one field in pandas?


```python
covid_Data_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
covid = pd.read_csv(covid_Data_url)
covid.head()
```

```python
Covidlat = covid['Lat']
Covidlat
```

### How to select multiple fields in pandas?

```python
Covidlatlong = covid[['Lat','Long']]
Covidlatlong
```

### How to select specific field with specific rows in pandas?

```python
covidsel = covid.loc[[1,4],['Lat','Long']]
covidsel
```

### How to select a range of fields with specific rows in pandas?

```python
covidRange = covid.loc[1:4,['Lat','Long']]
covidRange
```

### How to apply filter of a specific field in pandas?

```python
covidfilter =covid[covid['Lat']>31]
covidfilter
```

### How to drop columns in Pandas?

```python
coviddrop = covid.drop(['Lat','Long'],axis = 1)
coviddrop
```
